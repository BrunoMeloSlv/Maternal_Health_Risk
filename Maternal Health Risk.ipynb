```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_perc(st, data):
    plt.figure(figsize=(10,3))
    g = sns.countplot( x = st, data = data, orient = 'h')
    g.set_ylabel('Contagem', fontsize = 10)
    sizes = []

    for p in g.patches:
        height = p.get_height()
        sizes.append(height)
        g.text(p.get_x() + p.get_width()/1.6,
               height + 50,
               '{:1.2f}%'.format(height/data.shape[0] * 100),
               ha = 'center', va = 'bottom', fontsize = 12)
        g.set_ylim(0, max(sizes)*1.1)
```

## Transformando a base

```python
#Criando uma varivel 'disturbio do sono' a partir da varivel 'RiskLevel',
#com labels iguais a 0, 1 e 2 e com tipo 'int' (poderia tambm ser do tipo
#'float'), a fim de que seja possvel estimar o modelo por meio
#da funo 'MNLogit'
dados.loc[dados['RiskLevel']==
                            'low risk',
                            'Risco'] = 0 #categoria de referncia
dados.loc[dados['RiskLevel']==
                            'mid risk',
                            'Risco'] = 1
dados.loc[dados['RiskLevel']==
                            'high risk',
                            'Risco'] = 2

dados['Risco'] =\
    dados['Risco'].astype('int64')
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

image = img.imread('modelo.jpg')
plt.imshow(image)
plt.show()
```

```python
import statsmodels.api as sm # biblioteca de modelagem estatstica
import numpy as np # biblioteca para operaes matemticas multidimensionais
from scipy import stats # estatstica chi2
from statsmodels.iolib.summary2 import summary_col # comparao entre modelos
import plotly.graph_objs as go # grfico 3D
import statsmodels.formula.api as smf # estimao do modelo logstico binrio
from statsmodels.discrete.discrete_model import MNLogit
```

```python
x = dados.drop(columns=['RiskLevel','Risco'])
y = dados['Risco']

#Esse pacote precisa que a constante seja definida pelo usurio
X = sm.add_constant(x)

#Estimao do modelo - funo 'MNLogit' do pacote
#'statsmodels.discrete.discrete_model'
modelo_health = MNLogit(endog=y, exog=X).fit()

#Parmetros do modelo
modelo_health.summary()
```

```python
# In[ ]: Vamos definir uma funo 'Qui2' para se extrair a estatstica geral
# do modelo

def Qui2(modelo_multinomial):
    maximo = modelo_multinomial.llf
    minimo = modelo_multinomial.llnull
    qui2 = -2*(minimo - maximo)
    pvalue = stats.distributions.chi2.sf(qui2,1)
    df = pd.DataFrame({'Qui quadrado':[qui2],
                       'pvalue':[pvalue]})
    return df
```
